# BAB 2 TUTORIAL MATPLOTLIB PEMULA 

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Bikin data acak berantai (kumulatif) biar trennya naik-turun kayak saham
data = np.random.randn(50,).cumsum()
```

```python
# 1. membuat kanvas kosong
fig = plt.figure()

# 2. Bagi area jadi grid 2x2, lalu plot kertas di slot 1, 2, dan 3 (BaKoS)
ax1 = fig.add_subplot(2,2,1) # Kiri Atas
ax2 = fig.add_subplot(2,2,2) # Kanan Atas
ax3 = fig.add_subplot(2,2,3) # Kiri Bawah (Slot 4 Kanan Bawah kosong)

# 4. Gambar datanya di tiap kertas dengan style berbeda
ax1.plot(data, 'k--') # k=Hitam, -- = putus-putus
ax2.plot(data, 'r-')  # r=Merah, - = garis lurus
ax3.plot(data, 'b')   # b=Biru, (tanpa simbol = garis lurus)

plt.show() #menampilkan grafik

```

# WARNA, TANDA, & GAYA GARIS

```python
# Plot dengan **label='Default'** — label dipake buat legend nanti.
plt.plot(data, 'k--', label='Default')

plt.plot(data, 
         'k-', 
         drawstyle='steps-post', 
         label='Stepwise')

plt.legend() #memberi nama / tanda
```

### Sama kayak atas plus plt.legend() buat munculin kotak legenda di lokasi terbaik (loc='best').

```python
plt.plot(data, 'k-', 
         drawstyle='steps-post', 
         label='Stepwise')
plt.legend(loc='best')
```

# Ketebalan, Label dan Legenda

```python
fig = plt.figure()  
ax = fig.add_subplot(1,1,1)

# 1. Gambar grafik utama dari data kamu
ax.plot(data)

# 2. Tentukan posisi "paku" pembatas di sumbu X (angka mana aja yg mau dimunculin)
location = ax.set_xticks([0, 10, 20, 30, 40, 50])

# 3. Ganti angka "paku" tadi jadi tulisan, miringkan 30 derajat biar gak tabrakan
label = ax.set_xticklabels(
    ["one", "two", "three", "four", "five", "six"], 
    rotation=30, 
    fontsize="small"
)

# 4. Kasih nama judul grafik, label sumbu X (bawah), dan sumbu Y (kiri)
ax.set_title('First Graph')
ax.set_xlabel('Stages')
ax.set_ylabel('Anything')
```
