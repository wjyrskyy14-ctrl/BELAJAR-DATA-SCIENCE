### FEATURE ENGINEERING : Date Of Journey

(Feature Engineering adalah proses mengubah, memecah, atau menggabungkan kolom data mentah yang kita punya menjadi kolom baru yang lebih bermakna supaya model Machine Learning nantinya bisa belajar dengan lebih mudah dan akurat.)

Split kolom Date_of_Journey jadi 3 kolom baru: Date, Month, Year. Ini contoh klasik datetime feature engineering.

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.head(2)
```

```python
df.drop('Date_of_Journey', axis=1, inplace=True) # menghapus kolom date of journey. karna udh di ganti tdi

# axis=1 (Mau hapus Baris atau Kolom?) = 0 utk baris, 1 utk kolom

# inplace=True (Mau langsung disimpan atau cuma numpang lewat?) = True = hapus permanen
                                                                # false = nnti muncul lagi
```

```python
df.head()
```

```python
# Coba split Arrival_Time dulu — ambil bagian sebelum spasi (kasus '22:20 22 Mar', kita ambil '22:20').
df['Arrival_Time'].str.split(' ').str[0]
```

```python
# kita pecah Arrival_Time nya menjadi jam dan menit 
df['Arrival_Hour'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
df['Arrival_Min'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]
df.drop("Arrival_Time", axis=1, inplace=True)
```

```python
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Min'] = df['Arrival_Min'].astype(int)
```

```python
# cek struktur data nya setelah di FE
df.info()
```

```python
df.head(3)
```

```python
# Dep_Time dipecah jadi Dept_hour & Dept_min dalam 1 cell ringkas.
df['Dept_hour']=df['Dep_Time'].str.split(':').str[0]
df['Dept_min']=df['Dep_Time'].str.split(':').str[1]
df['Dept_hour']=df['Dept_hour'].astype(int)
df['Dept_min']=df['Dept_min'].astype(int)
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
# Drop kolom Route — gak relevan buat prediksi harga.
df.drop('Route',axis=1,inplace=True)
```

### One-Hot Encoding

One-Hot Encoding adalah proses mengubah data teks (kategori) menjadi angka biner (0 dan 1) berbentuk kolom-kolom baru, supaya model Machine Learning bisa membacanya tanpa salah paham.

```python
# Import OneHotEncoder dari sklearn.
from sklearn.preprocessing import OneHotEncoder
```

```python
# Inisialisasi encoder.
encoder = OneHotEncoder()
```

```python
# Fit + transform 3 kolom kategorikal jadi sparse matrix lalu convert ke array dense.
encoder.fit_transform(df[['Airline','Source', 'Destination']]).toarray()
```

```python
# Wrap hasil one-hot encoding jadi DataFrame dengan nama kolom otomatis dari get_feature_names_out().
pd.DataFrame(encoder.fit_transform(df[['Airline', 'Source', 'Destination']]).toarray(), columns=encoder.get_feature_names_out())
```
